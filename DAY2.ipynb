```python
from sklearn.linear_model import LinearRegression

x = [[1],[2],[3],[4],[5]]

y = [10,12,14,18,20]

model = LinearRegression()

model.fit(x, y)

prediction = model.predict([[6]])

print(f"Prediction value is: {prediction[0]}")
```

```python
from sklearn.cluster import KMeans

import numpy as np

# graph
x = np.array([[1,2],[1,4],[1,0],[10,2],[10,4],[10,0],[11,0]])

kmeans = KMeans(n_clusters = 3, random_state=0)

kmeans.fit(x)

#output

print(f"Clustering will be: \n {kmeans.cluster_centers_}")
```

# KMEANS ON IRIS DATASET

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.cluster import KMeans

#datasets
iris_df = datasets.load_iris()
#model
model = KMeans(n_clusters = 3)

#fitting
X = iris_df.data
model.fit(X)

iris_df = pd.DataFrame(data=iris_df.data, columns=iris_df.feature_names)
iris_df['cluster'] = model.labels_

#predicting
predicted_label = model.predict([[4, 2, 0.8, 1.6]])

#predicting on all the value present in the dataset
all_prediction = model.predict(X)

print(predicted_label)
print(all_prediction)
```

```python
iris_df.head()
```

```python
print(dir(iris_df))

import seaborn as sns

plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'sepal length (cm)', y = 'sepal width (cm)', hue ='cluster' , data = iris_df, palette = 'Set1')
plt.title('K-Means Clustering on Iris Dataset')
plt.show()
```

##K-MEANS CLUSTERING

```python
from sklearn import datasets

#loading dataset
df_cluster = pd.read_csv('kmeans.csv')
```

```python
df_cluster.head()
```

```python
from sklearn.cluster import KMeans

#model
model = KMeans(n_clusters = 3)

model.fit(df_cluster)
```

```python
# assigning cluster labels to dataframe
df_cluster['cluster'] = model.labels_


#predicting on new data
predicted_label = model.predict([[6.2, -4.2]])

#prediction on original data(without cluter labels)
all_prediction = model.predict(df_cluster[['Feature 1', 'Feature 2']])

print(df_cluster['cluster'])
print(all_prediction)
print(f"Predcited cluster for new data point: {predicted_label[0]}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Feature 1', y='Feature 2', hue='cluster', data=df_cluster, palette='Set2')
plt.title('K-Means Clustering on Dataset')
plt.show()
```
